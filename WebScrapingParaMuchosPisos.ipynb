```python
listaColumns =["ID","titulo","localizacion","precio","caracteristicas","caracteristicas2","calle","barrio","vendedor"]
diccionariosPiso = dict()
def RecogerDatosPiso(id):
    url = f"https://www.idealista.com/inmueble/{id}/"
    browser.get(url)
    time.sleep(random.randint(2,5))
    try:
        browser.find_element("xpath", '//*[@id="didomi-notice-agree-button"]').click()
    except:
        pass
    html = browser.page_source
    soup = bs(html,'lxml')
    titulo = soup.find("span",{"class":"main-info__title-main"}).text
    localizacion = soup.find("span",{"class":"main-info__title-minor"}).text.split(",")[0]
    precio = int(soup.find("span",{"class":"txt-bold"}).text.replace(".",""))
    todas_caracteristicas = soup.find("div",{"class":"details-property"}).find("div",{"class":"details-property-feature-one"})
    caracteristicas = [caracteristica.text.strip() for caracteristica in todas_caracteristicas.find_all("li")]
    todas_caracteristicas2 = soup.find("div",{"class":"details-property"}).find("div",{"class":"details-property-feature-two"})
    caracteristicas2 = [caracteristica.text.strip() for caracteristica in todas_caracteristicas2.find_all("li")]
    infoUbi = soup.find("div",{"id":"headerMap"})
    calle = infoUbi.find_all("li")[0].text.strip()
    barrio = infoUbi.find_all("li")[1].text.strip()
    vendedor = soup.find("div",{"class":"professional-name"}).find("span").text.strip()
    diccionariosPiso[id]=[int(id),titulo,localizacion,precio,caracteristicas,caracteristicas2,calle,barrio,vendedor]
```

**Aqui he hecho la prueba solo con dos indices lo suyo es poner todo lo que nos trae lo de arriba**

```python
for id in ids:
    RecogerDatosPiso(id)
```
